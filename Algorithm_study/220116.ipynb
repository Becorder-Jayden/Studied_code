```python
# q에 append 하지 않으면 while이 반복되지 않아..

from collections import deque

N, K = map(int, input().split())
MAX = 100000
visited = [0] * MAX

def bfs(n, k):
    q = deque()
    q.append(n)
    
    while q:
        pop = q.popleft()
        
        if pop == k:
            return visited[pop]
        
        for i in [2*pop, pop-1, pop+1]:
            if 0<=i<=MAX and not visited[i]:
                visited[i] = visited[pop] + 1
                q.append(i)
         
print(bfs(N, K))
```
