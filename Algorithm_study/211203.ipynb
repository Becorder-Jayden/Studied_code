```python
def prime(x):
    if x == 1:
        return False
    elif x in [2, 3]:
        return True
    else:
        for div in range(2, int(x**0.5)+1):
            if x % div == 0:
                return False
            return True

x = int(input())
prime(x)
```

```python
# N이 주어졌을 때 N/2를 찾고, N/2와 가장 가까운(+-) 소수 두 개를 찾는다
# 소수 = 2~n**0.5 범위를 확인했을 때, 합성수가 아닌 수

# 소수에 대해서 먼저 정의
def prime(N):
    for i in range(2, int(N**0.5)+1):
        for j in range(2, i):
            if i%j == 0:
                return '합성수'
            else:
                return '소수'

prime(12)
    
        

# N = int(input())

```
