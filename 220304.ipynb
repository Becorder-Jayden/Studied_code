```python
# Silver IV
# 숫자 카드__정답
```

```python
N = int(input())
has = list(map(int, input().split()))
has.sort()
M = int(input())
cards = list(map(int, input().split()))

def bs(li, target):
    start = 0
    end = len(li)-1
    while start <= end:
        mid = (start+end) // 2
        if li[mid] == target:
            return 1
        elif li[mid] < target:
            start = mid+1
        elif li[mid] > target:
            end = mid-1
    return 0
            
for i in cards:
    print(bs(has, i), end = ' ')
```
